```python
#Read dataset, check for null and duplicates
def read_dataset(dataset):
    
    data = pd.read_csv(dataset)
    nrow = len(data.index)
    ncol = len(data.columns)

    print("The dataset contains", format(nrow, ",d"), "rows and", ncol, "columns.")
    
    #Check for null values
    if ((data.isna().sum()).sum()) > 0:
        print("There are null items in the dataset")
    else:
        print("There are no null items in the dataset")
        
    #Check for duplicates
    
    #col_names = ["Country", "Year", "Status"]
    #(data.duplicated(subset=col_names)).sum()

    if (data.duplicated().sum()) > 0:
        print("There are duplicates in the dataset")
    else:
        print("There are no duplicates in the dataset")
        
    return data



#Function to categorize data into numeric and categorical
def categorize_data(data):
    
    numeric=[]
    categorical=[]
    numeric_dtypes = ["int64", "int32", "float64", "float32"]

    for i in range (len(data.columns)):
        if data[data.columns[i]].dtype in numeric_dtypes:
            numeric.append(data.columns[i])
        else:
            categorical.append(data.columns[i])
            
    return numeric, categorical

    
#Function to check for outliers
def outliers_check(data, numeric_cols):
    outliers_sum =[]

    for col in (numeric_cols):
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = (data[col] < (Q1 - 5 * IQR)) | (data[col] > (Q3 + 5 * IQR))
        print(col, "", outliers.sum())
        outliers_sum.append(outliers.sum())

    return outliers.sum()


def remove_duplicates(data):
    duplicated_sum = data.duplicated().sum()
    if duplicated_sum == 0:
        print("Number of duplicated rows in dataset =", duplicated_sum)
        return data
    else:
        print("Number of duplicated rows in dataset =", duplicated_sum)
        data = data[~data.duplicated()]
        print("Duplicated rows have been removed")
        return data

    
def remove_outliers(data, numeric_cols):
    
    for col in (numeric_cols):
        median_value = np.median(data[col])
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = (data[col] < (Q1 - 5 * IQR)) | (data[col] > (Q3 + 5 * IQR))
        #print(col, "", outliers.sum())
        data.loc[outliers, col] = median_value
    return data


# remove special characters from columns
def remove_spec_chars(data, categorical_cols):
    for col in categorical_cols:
        data[col] = data[col].str.replace(r'\W+', '').str.strip() #replaces special characters with white sapaces and removes the white spaces
    return data


def replace_unknown(data, categorical_cols):
    for col in categorical_cols:
        if "other" in data[col].values:
            #source: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.mode.html
            mode = data[col].mode()[0]
            data[col] = data[col].replace("unknown", mode)
    return data

    
def oneHotEncoding(data, categorical, drop_first):

    data_final = pd.get_dummies(data, columns=categorical, drop_first=drop_first)

    return data_final
```

```python
def predict(models, names, x_list, Y, eval_type):
    
    print(f"\033[1mEvaluating {eval_type} data\033[0m \n")
    
    fpr_list, tpr_list, roc_auc_list = [], [], []
    
    for model, name, X in zip(models, names, x_list):
        print(f"\033[1mEvaluating {name}...\033[0m")
        y_pred = model.predict(X)
        accuracy = accuracy_score(Y, y_pred)
        report = classification_report(Y, y_pred)
        print(f"{name} Accuracy:", accuracy)
        
        
        conf_matrix = confusion_matrix(Y, y_pred)
        plt.figure(figsize=(3, 2))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
        plt.title('Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
        print(f"{name} Classification report:\n", report)

    
        y_prob = model.predict_proba(X)[:, 1]
        fpr, tpr, _ = roc_curve(Y, y_prob)
        fpr_list.append(fpr)
        tpr_list.append(tpr)
        
        roc_auc = auc(fpr, tpr)
        #print(f"{name} AUC = {roc_auc}")
        roc_auc_list.append(roc_auc)
        
    cmap = plt.colormaps['tab10']
    colors = cmap.colors[:len(names)]
    plt.figure(figsize=(7,5))
    for i, (fpr, tpr, roc_auc, name) in enumerate(zip(fpr_list, tpr_list, roc_auc_list, names)):
        plt.plot(fpr, tpr, color=colors[i], lw=2, label=f'{name} (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC)')
    plt.legend(loc='lower right')
    plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
df = pd.DataFrame(data_sampled)
# Define feature matrix X and target vector Y
X = df.drop('Revenue_True', axis=1)
Y = df['Revenue_True']

# Identify numeric columns for scaling
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns

# Ensure both classes are present
print("Class distribution in the entire dataset:")
print(Y.value_counts())

# Stratified split to maintain class distribution
x, xtest, y, ytest = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
xtrain, xval, ytrain, yval = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)

# Check class distribution
print("\nClass distribution after stratified splitting:")
print("Training set:\n", ytrain.value_counts())
print("Validation set:\n", yval.value_counts())
print("Test set:\n", ytest.value_counts())

# Undersampling
try:
    undersampling = RandomUnderSampler(random_state=42)
    xtrain_undersampled, ytrain_undersampled = undersampling.fit_resample(xtrain, ytrain)
    print("\nClass distribution after undersampling:")
    print(ytrain_undersampled.value_counts())
except ValueError as e:
    print("\nError during undersampling:", e)

# Oversampling
try:
    oversampling = RandomOverSampler(random_state=42)
    xtrain_oversampled, ytrain_oversampled = oversampling.fit_resample(xtrain, ytrain)
    print("\nClass distribution after oversampling:")
    print(ytrain_oversampled.value_counts())
except ValueError as e:
    print("\nError during oversampling:", e)

# Scaling the data - oversampled dataset & undersampled
scaler = StandardScaler()
xtrain_oversampled[numeric_cols] = scaler.fit_transform(xtrain_oversampled[numeric_cols])
xtrain_undersampled[numeric_cols] = scaler.fit_transform(xtrain_undersampled[numeric_cols])
xtest[numeric_cols] = scaler.transform(xtest[numeric_cols])
xval[numeric_cols] = scaler.transform(xval[numeric_cols])

# PCA - oversampled dataset
pca_oversampled = PCA(n_components=2)
xtrain_pca_oversampled = pca_oversampled.fit_transform(xtrain_oversampled)
xval_pca_oversampled = pca_oversampled.transform(xval)
xtest_pca_oversampled = pca_oversampled.transform(xtest)

# PCA - undersampled dataset
pca_undersampled = PCA(n_components=2)
xtrain_pca_undersampled = pca_undersampled.fit_transform(xtrain_undersampled)
xval_pca_undersampled = pca_undersampled.transform(xval)
xtest_pca_undersampled = pca_undersampled.transform(xtest)

# Define the parameter grid for Grid Search
param_grid = {
    'hidden_layer_sizes': [(5,), (10,), (20,), (5, 5), (10, 10), (20, 20)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant', 'adaptive'],
}

best_accuracy = 0
best_random_state = None

for random_state in range(101):
    # Stratified split to maintain class distribution
    x, xtest, y, ytest = train_test_split(X, Y, test_size=0.2, random_state=random_state, stratify=Y)
    xtrain, xval, ytrain, yval = train_test_split(x, y, test_size=0.3, random_state=random_state, stratify=y)

    # Undersampling
    undersampling = RandomUnderSampler(random_state=random_state)
    xtrain_undersampled, ytrain_undersampled = undersampling.fit_resample(xtrain, ytrain)

    # Oversampling
    oversampling = RandomOverSampler(random_state=random_state)
    xtrain_oversampled, ytrain_oversampled = oversampling.fit_resample(xtrain, ytrain)

    # Scaling the data - oversampled dataset & undersampled
    scaler = StandardScaler()
    xtrain_oversampled[numeric_cols] = scaler.fit_transform(xtrain_oversampled[numeric_cols])
    xtrain_undersampled[numeric_cols] = scaler.fit_transform(xtrain_undersampled[numeric_cols])
    xtest[numeric_cols] = scaler.transform(xtest[numeric_cols])
    xval[numeric_cols] = scaler.transform(xval[numeric_cols])

    # PCA - oversampled dataset
    pca_oversampled = PCA(n_components=2)
    xtrain_pca_oversampled = pca_oversampled.fit_transform(xtrain_oversampled)
    xval_pca_oversampled = pca_oversampled.transform(xval)
    xtest_pca_oversampled = pca_oversampled.transform(xtest)

    # PCA - undersampled dataset
    pca_undersampled = PCA(n_components=2)
    xtrain_pca_undersampled = pca_undersampled.fit_transform(xtrain_undersampled)
    xval_pca_undersampled = pca_undersampled.transform(xval)
    xtest_pca_undersampled = pca_undersampled.transform(xtest)

    # Grid Search on undersampled data
    mlp_undersampled = MLPClassifier(random_state=random_state, max_iter=1000)
    grid_search_undersampled = GridSearchCV(estimator=mlp_undersampled, param_grid=param_grid, n_jobs=-1, cv=3, scoring='accuracy')
    grid_search_undersampled.fit(xtrain_pca_undersampled, ytrain_undersampled)

    # Best parameters and model for undersampled data
    best_params_undersampled = grid_search_undersampled.best_params_
    best_model_undersampled = grid_search_undersampled.best_estimator_

    yval_pred_undersampled = best_model_undersampled.predict(xval_pca_undersampled)
    ytest_pred_undersampled = best_model_undersampled.predict(xtest_pca_undersampled)

    # Grid Search on oversampled data
    mlp_oversampled = MLPClassifier(random_state=random_state, max_iter=1000)
    grid_search_oversampled = GridSearchCV(estimator=mlp_oversampled, param_grid=param_grid, n_jobs=-1, cv=3, scoring='accuracy')
    grid_search_oversampled.fit(xtrain_pca_oversampled, ytrain_oversampled)

    # Best parameters and model for oversampled data
    best_params_oversampled = grid_search_oversampled.best_params_
    best_model_oversampled = grid_search_oversampled.best_estimator_

    yval_pred_oversampled = best_model_oversampled.predict(xval_pca_oversampled)
    ytest_pred_oversampled = best_model_oversampled.predict(xtest_pca_oversampled)

    # Evaluation - Oversampled
    accuracy = accuracy_score(ytest, ytest_pred_oversampled)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_random_state = random_state

print(f"The best random state is {best_random_state} with an accuracy of {best_accuracy}")

# Re-evaluate the best random state
x, xtest, y, ytest = train_test_split(X, Y, test_size=0.2, random_state=best_random_state, stratify=Y)
xtrain, xval, ytrain, yval = train_test_split(x, y, test_size=0.3, random_state=best_random_state, stratify=y)

# Undersampling
undersampling = RandomUnderSampler(random_state=best_random_state)
xtrain_undersampled, ytrain_undersampled = undersampling.fit_resample(xtrain, ytrain)

# Oversampling
oversampling = RandomOverSampler(random_state=best_random_state)
xtrain_oversampled, ytrain_oversampled = oversampling.fit_resample(xtrain, ytrain)

# Scaling the data - oversampled dataset & undersampled
scaler = StandardScaler()
xtrain_oversampled[numeric_cols] = scaler.fit_transform(xtrain_oversampled[numeric_cols])
xtrain_undersampled[numeric_cols] = scaler.fit_transform(xtrain_undersampled[numeric_cols])
xtest[numeric_cols] = scaler.transform(xtest[numeric_cols])
xval[numeric_cols] = scaler.transform(xval[numeric_cols])

# PCA - oversampled dataset
pca_oversampled = PCA(n_components=2)
xtrain_pca_oversampled = pca_oversampled.fit_transform(xtrain_oversampled)
xval_pca_oversampled = pca_oversampled.transform(xval)
xtest_pca_oversampled = pca_oversampled.transform(xtest)

# PCA - undersampled dataset
pca_undersampled = PCA(n_components=2)
xtrain_pca_undersampled = pca_undersampled.fit_transform(xtrain_undersampled)
xval_pca_undersampled = pca_undersampled.transform(xval)
xtest_pca_undersampled = pca_undersampled.transform(xtest)

# Grid Search on undersampled data
mlp_undersampled = MLPClassifier(random_state=best_random_state, max_iter=1000)
grid_search_undersampled = GridSearchCV(estimator=mlp_undersampled, param_grid=param_grid, n_jobs=-1, cv=3, scoring='accuracy')
grid_search_undersampled.fit(xtrain_pca_undersampled, ytrain_undersampled)

# Best parameters and model for undersampled data
best_params_undersampled = grid_search_undersampled.best_params_
best_model_undersampled = grid_search_undersampled.best_estimator_

yval_pred_undersampled = best_model_undersampled.predict(xval_pca_undersampled)
ytest_pred_undersampled = best_model_undersampled.predict(xtest_pca_undersampled)

# Grid Search on oversampled data
mlp_oversampled = MLPClassifier(random_state=best_random_state, max_iter=1000)
grid_search_oversampled = GridSearchCV(estimator=mlp_oversampled, param_grid=param_grid, n_jobs=-1, cv=3, scoring='accuracy')
grid_search_oversampled.fit(xtrain_pca_oversampled, ytrain_oversampled)

# Best parameters and model for oversampled data
best_params_oversampled = grid_search_oversampled.best_params_
best_model_oversampled = grid_search_oversampled.best_estimator_

yval_pred_oversampled = best_model_oversampled.predict(xval_pca_oversampled)
ytest_pred_oversampled = best_model_oversampled.predict(xtest_pca_oversampled)

# Evaluation - Undersampled
print("Best parameters for undersampled data:", best_params_undersampled)
print("\nUndersampled Data - Validation Set")
print(classification_report(yval, yval_pred_undersampled))
print("Confusion Matrix:\n", confusion_matrix(yval, yval_pred_undersampled))
print("Accuracy:", accuracy_score(yval, yval_pred_undersampled))

print("\nUndersampled Data - Test Set")
print(classification_report(ytest, ytest_pred_undersampled))
print("Confusion Matrix:\n", confusion_matrix(ytest, ytest_pred_undersampled))
print("Accuracy:", accuracy_score(ytest, ytest_pred_undersampled))

# Evaluation - Oversampled
print("Best parameters for oversampled data:", best_params_oversampled)
print("\nOversampled Data - Validation Set")
print(classification_report(yval, yval_pred_oversampled))
print("Confusion Matrix:\n", confusion_matrix(yval, yval_pred_oversampled))
print("Accuracy:", accuracy_score(yval, yval_pred_oversampled))
# Evaluation - Oversampled
print("Best parameters for oversampled data:", best_params_oversampled)
print("\nOversampled Data - Validation Set")
print(classification_report(yval, yval_pred_oversampled))
print("Confusion Matrix:\n", confusion_matrix(yval, yval_pred_oversampled))
print("Accuracy:", accuracy_score(yval, yval_pred_oversampled))

print("\nOversampled Data - Test Set")
print(classification_report(ytest, ytest_pred_oversampled))
print("Confusion Matrix:\n", confusion_matrix(ytest, ytest_pred_oversampled))
print("Accuracy:", accuracy_score(ytest, ytest_pred_oversampled))
```

# Conclusion

The imbalanced nature of the dataset poses a significant challenge for the logistic regression model. Class imbalance is particularly evident in the "No Sampling" scenario, where the model tends to favor the majority class.
Both undersampling and oversampling techniques aim to address class imbalance, but their impact on model performance is not positive in this case. The reduced accuracy and suboptimal confusion matrices suggest that the logistic regression model may not be well-suited for capturing the underlying patterns in the data.
Further exploration of more sophisticated algorithms, hyperparameter tuning, or alternative sampling techniques may be necessary to improve model performance.
